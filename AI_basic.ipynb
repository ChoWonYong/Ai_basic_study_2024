```python
import sys
sys.path.append('/content/drive/MyDrive/number')
import numpy as np
from PIL import Image
from AImod import *

I_0, I_1, I_2, I_3, I_4, I_5, I_6, I_7, I_8, I_9 = [], [], [], [], [], [], [], [], [], []

j_0, j_1, j_2, j_3, j_4, j_5, j_6, j_7, j_8, j_9 = [], [], [], [], [], [], [], [], [], []

# train, test dataset의 원소 I_i, j_i들은 숫자i의 사진 list들
train_dataset = [I_0, I_1, I_2, I_3, I_4, I_5, I_6, I_7, I_8, I_9]
test_dataset = [j_0, j_1, j_2, j_3, j_4, j_5, j_6, j_7, j_8, j_9]

train_labels = []
test_labels = []

for i in range(10):
    one_hot = np.zeros(10)
    one_hot[i] = 1

    train_labels.append(one_hot)

    test_labels.append(one_hot)

loading_image = 10

def imagetomatrix(k, text):
    count, i = 0, 0
    while count < loading_image:
        try:
            image_path = f"{text}/{k}/{i}.png"
            # 해당 경로의 i.png를 탐색함 만약 i.png가 없으면 error가 발생하고, 예외처리가 되어 아래의 except:가 실행됨.
            img = Image.open(image_path)
            img_np = np.array(img)
            a = img_np.flatten()
            b = (a, k)
            if text == 'train':
                train_dataset[k].append(b)
            elif text == 'valid':
                test_dataset[k].append(b)
            count += 1  # 한장 찾았으니 count를 1 늘림
        except FileNotFoundError:
            pass  # pass는 그냥 넘긴다는 뜻. 파일을 찾지 못해 에러가 발생했으니 그냥 넘어가는 것.
        i += 1  # 찾건 못찾건 i는 증가함. 다음 순서의 파일이 있는지 확인하는 것


# train_dataset의 원소 I_k에 이미지의 행렬정보 넣기
for i in range(10):
    imagetomatrix(i, 'train')
# test_dataset의 원소 j_k에 이미지의 행렬정보 넣기
for i in range(10):
    imagetomatrix(i, 'valid')
```

```python
class Neural_Net:
    def __init__(self, layer_size):  # 클래스가 갖는 속성을 설정.
        self.layer_number = len(layer_size)
        self.W = ['Weight']
        for i in range(self.layer_number - 1):
          self.W.append(0.01 * np.random.randn(layer_size[i + 1], layer_size[i]))
        self.B = ['Bias']
        for i in range(self.layer_number - 1):
            self.B.append(0.01 * np.random.randn(layer_size[i + 1], 1))

    def forward_propagation(self, input):
        self.L = ['Unassigned'] * self.layer_number
        X = np.array(input).reshape(len(input), 1)
        self.L[0] = X
        self.L[1] = self.W[1] @ self.L[0] + self.B[1]
        for i in range(2, self.layer_number):
            self.L[i] = self.W[i] @ ReLU(self.L[i - 1]) + self.B[i]  # 각 layer값들을 리스트 L에 저장
        return softmax(self.L[self.layer_number - 1])

    def back_propagation(self, x, T):
        Y = self.forward_propagation(x)
        Y = np.array(Y)
        T = np.array(T)
        for i in range(len(self.W) - 1):
            self.W[i + 1] -= cal_pNpW(self.W, self.L, Y, T)[i].reshape(self.W[i + 1].shape) * step_size  # 왜 i+1인지 생각해볼 것
            self.B[i + 1] -= cal_pNpB(self.W, self.L, Y, T)[i].reshape(self.B[i + 1].shape) * step_size

step_size = 0.1
NN = Neural_Net([784, 256, 128, 64, 10])

iteration = 1000
for _ in range(iteration):
    for j in range(loading_image):
        for i in range(10):
            NN.back_propagation(train_dataset[i][j][0], train_labels[i])
    error = 0
    for i in range(10):
        for j in range(loading_image):
            error += MSE(NN.forward_propagation(test_dataset[i][j][0]), test_labels[i])
    print('현재 오차는:', error / (10 * loading_image))
```
